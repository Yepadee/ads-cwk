```python

from gensim.parsing.preprocessing import STOPWORDS
import gensim
from nltk.stem import WordNetLemmatizer
import pyLDAvis
import pyLDAvis
import pyLDAvis.gensim
import pandas as pd
```

```python

def preprocess(text):
    result=[]
    for token in gensim.utils.simple_preprocess(text) :
        if token not in gensim.parsing.preprocessing.STOPWORDS:
            result.append(WordNetLemmatizer().lemmatize(token, 'v')) # for now, lemmatizing only verbs
    return result


def build_lda_model(reviews):
    processed = [preprocess(text) for text in reviews]
    dictionary = gensim.corpora.Dictionary(processed) # construct word<->id mappings - it does it in alphabetical order
    bow_corpus = [dictionary.doc2bow(review) for review in processed]
    lda_model =  gensim.models.LdaModel(bow_corpus, 
                                       num_topics=3, 
                                       id2word=dictionary,                                    
                                       passes=10,
                                       )
    return (lda_model, bow_corpus, dictionary)
```

```python
def visualize_lda(lda_model, bow_corpus, dictionary):
    pyLDAvis.enable_notebook()
    LDAvis_prepared = pyLDAvis.gensim.prepare(lda_model, bow_corpus, dictionary)
    return LDAvis_prepared 

```

```python
df = pd.read_csv("bad-reviews.csv")
```

```python
df.head()
```

```python
reviews = df["review_body"]
```

```python
lda, bow_corpus, dictionary = build_lda_model(reviews)
```

```python
visualize_lda(lda, bow_corpus, dictionary)
```
